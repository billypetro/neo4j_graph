### Creating Relationships

      - here you are sying that Tom Hanks acted in Castaway. Or that any node has a relationship with the other. we difine the direction simply with the arrow and which way it points
  
  Here we will do the same with Michael Cain and Batman

By default, the graph engine will create the relationship from left to right between the nodes, or you could specify the relationship with an arrowwe can create a relationship where the fields exist like this

# this will CREATE the movie, then create the relationship in one go!
    merge((m:Movie {name:'Black Panther'})) #creates the Movie
